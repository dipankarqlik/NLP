```python
# FUNCTION TO PULL THE REQUIRED COLUMNS FROM XML INTO A DATAFRAME:

def myFunction(file):
    date_published=""
    tree = ET.parse(file)
    xml_data = tree.getroot()
    item_id = xml_data.attrib['itemid']
    headline = xml_data[1].text
    
    listx = []
    text = xml_data.find('text')
    for z in text:
        listx.append(z.text)
        for text in listx:
            text=" ".join(str(text))
    
    for x in xml_data.findall('metadata/dc'):
        if x.attrib['element']=="dc.date.published":
            date_published=x.attrib['value']
            
    
    bip_topics=[]
    getel= xml_data.find("metadata")
    for y in getel:
        if y.get("class")=="bip:topics:1.0":
            for z in y:
                bip_topics.append(z.get("code"))
    bip_topics= ",".join(bip_topics)
    
    XMLfilename=os.path.splitext(os.path.basename(file_path))[0]
    good_ds=[headline, item_id, text, date_published, bip_topics, XMLfilename]
    return good_ds
```

# FUNCTION TO TOKENIZE COLUMN 'HEADLINE':

```python
def function_to_clear(values):
    headline = values['headline']
    tokens = nltk.word_tokenize(headline)
    token_form = [x for x in tokens if x.isalpha()]
    return token_form

new_df['words'] = new_df.apply(function_to_clear, axis=1)
```

# IMPORT STOP WORDS

```python
#Import stop words from library:
from nltk.corpus import stopwords
stop_words_list = set(stopwords.words("english"))
```

# FUNCTION TO REMOVE STOP WORDS FROM ALL ROWS:

```python
def remove_stop_words(values):
    my_list = values['words']
    corrected_words = [x for x in my_list if not x in stop_words_list]
    return (corrected_words)

new_df['meaningful_words'] = new_df.apply(remove_stop_words, axis=1)
```

```python
def join_words(values):
    my_list = values['meaningful_words']
    newly_formed = ( " ".join(my_list))
    return newly_formed

new_df['pr_words'] = new_df.apply(join_words, axis=1)
```
